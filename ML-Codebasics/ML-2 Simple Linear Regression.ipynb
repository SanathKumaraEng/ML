```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price, color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']].values,df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price, color = 'red', marker = '+')
plt.plot(df.area,reg.predict(df[['area']].values),color='blue')
```

```python
d = pd.read_csv("areas.csv")
d
```

```python
p = reg.predict(d.values)
```

```python
d['price'] = p
```

```python
d
```

```python
d.to_csv("prediction.csv", index = False)
```

```python
canada_df = pd.read_csv("canada_per_capita_income.csv")
canada_df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(canada_df.year,canada_df.per_capita_income, color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(canada_df[['year']].values,canada_df.per_capita_income)
```

```python
reg.predict([[20500]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(canada_df.year,canada_df.per_capita_income, color = 'red', marker = '+')
plt.plot(canada_df.year,reg.predict(canada_df[['year']].values),color='blue')
```
